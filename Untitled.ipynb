### Fitting a Regression Model

```python
df
```

```python
np.sum(df.isna())
```

```python
mod_data = df.copy()
```

```python
mod_data['Prev_Close'] = mod_data['Close'].shift(1)
```

```python
mod_data = mod_data[["Close","Prev_Close"]]
```

```python
mod_data = mod_data.dropna()
```

```python
mod_data
```

```python
x_train = mod_data[["Prev_Close"]][:-1]; y_train = mod_data[["Close"]][:-1]
x_test = mod_data[["Prev_Close"]].iloc[[-1]]; y_test = mod_data[["Close"]].iloc[[-1]]
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
type(x_test)
```

```python
# 2) Model (tweak hyperparameters as needed)
gbr = GradientBoostingRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=3,
    subsample=0.9,
    random_state=42
)

# 3) Train
gbr.fit(x_train, y_train)

# 4) Predict the held-out last point
y_pred = gbr.predict(x_test)[0]

# 5) Quick evaluation on the training window (optional)
y_fit = gbr.predict(x_train)
mae  = mean_absolute_error(y_train, y_fit)
rmse = mean_squared_error(y_train, y_fit)
r2   = r2_score(y_train, y_fit)

print(f"Next/last prediction: {y_pred} (true: {y_test})")
print(f"Train MAE: {mae:.4f}, RMSE: {rmse:.4f}, R²: {r2:.4f}")
```

```python
mod_data['Predicted'] = gbr.predict(mod_data[['Prev_Close']])

plt.figure(figsize=(12,6))
plt.plot(mod_data['Close'], label='Actual', linewidth=2)
plt.plot(mod_data['Predicted'], label='Predicted', linestyle='--')
plt.title('Gradient Boosting Regressor — Actual vs Predicted')
plt.xlabel('Index')
plt.ylabel('Close Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot the full actual series
plt.figure(figsize=(12,6))
plt.plot(mod_data['Close'].values, label='Actual Close', color='blue', linewidth=2)

# Highlight the last actual point
plt.scatter(len(mod_data)-1, y_test, color='green', s=100, label='Actual (last point)')

# Add your model’s predicted point
plt.scatter(len(mod_data)-1, y_pred, color='red', s=100, marker='x', label='Predicted (last point)')

# Labels and formatting
plt.title('Actual Time Series with Last Predicted Point')
plt.xlabel('Time (index)')
plt.ylabel('Close Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from joblib import dump, load

# Save the trained model
dump(gbr, "gbr_model.joblib")

# Later (or in another notebook/script), load it again
gbr_loaded = load("gbr_model.joblib")

# Test it
y_pred = gbr_loaded.predict(x_test)
```
